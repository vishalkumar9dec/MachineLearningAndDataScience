```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
digits_data=load_digits()
```

```python
dir(digits_data)
```

```python
df_digits=pd.DataFrame(digits_data.data)
```

```python
df_digits.head()
```

```python
df_digits['target']=digits_data.target
```

```python
df_digits.head(15)
```

```python
X=df_digits.drop(columns=['target'])
y=df_digits['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_digits=RandomForestClassifier(n_estimators=20)
```

```python
model_digits.fit(X_train,y_train)
```

```python
model_digits.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,model_digits.predict(X_test))
```

```python
sns.heatmap(cm,annot=True)
```
