```python
from sklearn.feature_extraction.text import CountVectorizer
vector=CountVectorizer()
```

We will be using pipeline method here so that CountVectorizer and multinomialNB are fitted in one step

```python
from sklearn.pipeline import Pipeline
```

```python
pip_model=Pipeline([('vectorize',CountVectorizer()),('nb',MultinomialNB())])
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_mails['Message'],df_mails['SpamMail'],test_size=0.2)
```

```python
pip_model.fit(X_train,y_train)
```

```python
pip_model.predict(emails)
```

```python
pip_model.score(X_test,y_test)
```
